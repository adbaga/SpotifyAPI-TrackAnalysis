```python
for i in range(len(pl3.df)):
    pl3.df['sad'] = 1
for i in range(len(pl4.df)):
    pl4.df['sad'] = 0
```

```python
pl3.df.head()
```

```python
pl4.df.head()
```

```python
features = pl3.df.columns[6:]

pl3.avPl3 = []
pl3.stdPl3 = []
for i in features:
    av = pl3.df[i].mean()
    std = pl3.df[i].std()
    pl3.avPl3.append(av)
    pl3.stdPl3.append(std)


pl4.avPl4 = []
pl4.stdPl4 = []
for i in features:
    av = pl4.df[i].mean()
    std = pl4.df[i].std()
    pl4.avPl4.append(av)
    pl4.stdPl4.append(std)


print(features)
    
```

```python
pl34DFmean = pd.DataFrame([pl3.avPl3, pl4.avPl4],  columns =features)
```
